```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import geopandas as gpd
import statsmodels.api as sm
```

```python
df = pd.read_csv("sleepdata.csv")
df.head()
```

```python
df['GDP_per_Capita'] = df['GDP_per_Capita'].astype(str).str.replace(',', '').astype(float)
```

```python
import numpy as np
df['GDP_per_Capita'] = pd.to_numeric(df['GDP_per_Capita'], errors='coerce')
```

```python
# scatterplot graphing Bedtime vs GDP 
sns.set_theme(style='white', palette='Set2')

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=df, x="GDP_per_Capita",y="Bedtime", hue="Cultural_Group", ax=ax)

for i, row in df.iterrows():
    if row["Countries"] == "Switzerland":
        ax.annotate(row["Countries"], xy=(row["GDP_per_Capita"], row["Bedtime"]),
                    xytext=(-50, 6), textcoords='offset points', fontsize=7)
    else:
        ax.annotate(row["Countries"], xy=(row["GDP_per_Capita"], row["Bedtime"]),
                    xytext=(6, 4), textcoords='offset points', fontsize=7)

sns.regplot(data=df, x='GDP_per_Capita', y='Bedtime', scatter=False, ax=ax, color='gray', line_kws={'linestyle':'--'})

plt.title('Bedtime vs GDP per Capita by Cultural Group')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Bedtime (hour)')

ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style='white', palette='Set2')

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=df, x="Idv_BW",y="Bedtime", hue="Cultural_Group", ax=ax)

for i, row in df.iterrows():
    if row["Countries"]=="South Korea":
        ax.annotate(row["Countries"], xy=(row["Idv_BW"], row["Bedtime"]),
                    xytext=(-40, 4), textcoords='offset points', fontsize=7)
    elif row["Countries"]=="Germany":
        ax.annotate(row["Countries"], xy=(row["Idv_BW"], row["Bedtime"]),
                    xytext=(6, -6), textcoords='offset points', fontsize=7)
    elif row["Countries"]=="China":
        ax.annotate(row["Countries"], xy=(row["Idv_BW"], row["Bedtime"]),
                    xytext=(-25, 4), textcoords='offset points', fontsize=7)
    elif row["Countries"]=="Malaysia":
        ax.annotate(row["Countries"], xy=(row["Idv_BW"], row["Bedtime"]),
                    xytext=(6, -4), textcoords='offset points', fontsize=7)
    else:
        ax.annotate(row["Countries"], xy=(row["Idv_BW"], row["Bedtime"]),
                    xytext=(6, 4), textcoords='offset points', fontsize=7)


sns.regplot(data=df, x="Idv_BW", y='Bedtime', scatter=False, ax=ax, color='gray', line_kws={'linestyle':'--'})

plt.title("Bedtime vs Beugelsdijik and Welzel (BW) Individualism Index")
plt.xlabel("BW Individualism Index")
plt.ylabel('Bedtime (hour)')

ax.set_xlim(15, 73)  
ax.set_ylim(19, 23)  

ax.grid(True, linestyle='--', alpha=0.4)
# plt.tight_layout()
plt.show()
```

```python
# scatterplot graphing Beddtime vs Population Density 
sns.set_theme(style='white', palette='Set2')

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=df, x="Population_Density",y="Bedtime", hue="Cultural_Group", ax=ax)

for i, row in df.iterrows():
    if row["Countries"] == "Singapore":
        ax.annotate(row["Countries"], xy=(row["Population_Density"], row["Bedtime"]),
                    xytext=(-55, 4), textcoords='offset points', fontsize=7)
    elif row["Countries"]== "Germany":
        ax.annotate(row["Countries"], xy=(row["Population_Density"], row["Bedtime"]),
                    xytext=(-40, 4), textcoords='offset points', fontsize=7)
    else:
        ax.annotate(row["Countries"], xy=(row["Population_Density"], row["Bedtime"]),
                    xytext=(6, 4), textcoords='offset points', fontsize=7)

# ax.set_xscale('log')
ax.set_xscale('linear')

sns.regplot(data=df, x="Population_Density", y='Bedtime', scatter=False, ax=ax, color='gray', line_kws={'linestyle':'--'}, ci=None)

plt.title("Bedtime vs Population Density")
plt.xlabel("Population Density")
plt.ylabel('Bedtime (hour)')

ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="white", palette="Set2")

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=df, x="Idv_BW",y="Co-Sleeping_Percentage", hue="Cultural_Group", ax=ax)

for i, row in df.iterrows():
    ax.annotate(row["Countries"], xy=(row["Idv_BW"], row["Co-Sleeping_Percentage"]),
                    xytext=(6, 4), textcoords='offset points', fontsize=7)

# ax.set_xscale('log')
ax.set_xscale('linear')

sns.regplot(data=df, x="Idv_BW", y='Co-Sleeping_Percentage', scatter=False, ax=ax, color='gray', line_kws={'linestyle':'--'})

plt.title("Individualism vs Co-Sleeping")
plt.xlabel("Individualism Index")
plt.ylabel('Percentage of Infants who Co-Sleep')

ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
